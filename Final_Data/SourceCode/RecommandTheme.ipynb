```
def rgb2lab(R, G, B) :
  var_R = R/255.0;
  var_G = G/255.0;
  var_B = B/255.0;


  if var_R > 0.04045 :
    var_R = pow( (( var_R + 0.055 ) / 1.055 ), 2.4 )
  else :
    var_R = var_R / 12.92
  if var_G > 0.04045 :
    var_G = pow( ( ( var_G + 0.055 ) / 1.055 ), 2.4)
  else :
    var_G = var_G / 12.92
  if var_B > 0.04045 : 
    var_B = pow( ( ( var_B + 0.055 ) / 1.055 ), 2.4)
  else :
    var_B = var_B / 12.92

  var_R = var_R * 100.
  var_G = var_G * 100.
  var_B = var_B * 100.

  X = var_R * 0.4124 + var_G * 0.3576 + var_B * 0.1805
  Y = var_R * 0.2126 + var_G * 0.7152 + var_B * 0.0722
  Z = var_R * 0.0193 + var_G * 0.1192 + var_B * 0.9505


  var_X = X / 95.047         
  var_Y = Y / 100.000          
  var_Z = Z / 108.883

  if var_X > 0.008856 :
    var_X = pow(var_X , ( 1./3. ) )
  else :
    var_X = ( 7.787 * var_X ) + ( 16. / 116. )
  if var_Y > 0.008856 :
    var_Y = pow(var_Y , ( 1./3. ))
  else :
    var_Y = ( 7.787 * var_Y ) + ( 16. / 116. )
  if var_Z > 0.008856 :
    var_Z = pow(var_Z , ( 1./3. ))
  else :
    var_Z = ( 7.787 * var_Z ) + ( 16. / 116. )

  l_s = ( 116. * var_Y ) - 16.
  a_s = 500. * ( var_X - var_Y )
  b_s = 200. * ( var_Y - var_Z )

  return [l_s, a_s, b_s]
```

```
# inverse_distance
def inverse_distance(a,b) :
  if a[0] == b[0] and a[1] == b[1] and a[2] == b[2] : #동일한 색인 경우
    return 1.0 
  return 1/(sum((a-b)**2))**(1/2)
```

```
#colorChip과 유사도가 상위 10개 안에 드는 theme속의 컬러를 colorChip의 컬러로 치환

labThemeData = pd.read_csv('gdrive/My Drive/colorPOOL/kulerData/kulerLabData.csv')
labThemeData = labThemeData.iloc[:,1:]
labThemeData
colorChipLabData = pd.read_csv('gdrive/My Drive/colorPOOL/kulerData/ColorChipLabData.csv')
colorChipLabData = colorChipLabData.iloc[:,1:]
colorChipLabData
recommandTheme = pd.DataFrame(columns=['themeid1','themeid2','themeid3','themeid4','themeid5','themeid6','themeid7','themeid8','themeid9','themeid10']) # color칩에 대한 추천 테마의 id를 저장할까 아니면 Lab값을 다 저장할까?

visited = [[False for j in range(5)] for i in range(3000)] # 테마내의 컬러가 컬러칩내의 컬러로 치환되었을 때 방문 표시

for i in range(len(colorChipLabData)) : #len(colorChipLabData)
  themeRow = []# 테마 데이터의 행 좌표
  themeCol = []# 테마 데이터의 열 좌표(L좌표)
  themeSim = []# 테마 데이터의 유사도 
  L1 = colorChipLabData.iloc[i,0] #colorChip의 L값
  a1 = colorChipLabData.iloc[i,1] #colorChip의 a값
  b1 = colorChipLabData.iloc[i,2] #colorChip의 b값
  color1 = np.array([L1, a1, b1]) #colorChip[i]를 numpy 배열로 저장
  
  for j in range(len(labThemeData)) : #len(labThemeData)
    for k in range(5) :
      L2 = labThemeData.iloc[j,3*k] #theme의 L값
      a2 = labThemeData.iloc[j,3*k+1] #theme의 a값
      b2 = labThemeData.iloc[j,3*k+2] #theme의 b값
      color2 = np.array([L2,a2,b2]) #theme의 컬러를 numpy 배열로 변환
      sim = inverse_distance(color1, color2); #유사도(거리의 역수 측정)
      themeRow.append(j)
      themeCol.append(k)
      themeSim.append(sim)

  themeSimData = list(zip(themeRow,themeCol,themeSim)) # 튜플리스트로 저장
  themeSimData = sorted(themeSimData, key=lambda x : x[2], reverse=True) #유사도의 내립차순으로 정렬
  
  # 상위 10개의 유사도를 가지는 theme 컬러에 대해서 colorChip데이터로 치환
  count = 0
  recommandThemeOne = []
  for j in range(len(themeSimData)) : 
    row = themeSimData[j][0]
    col = themeSimData[j][1]
    sim = themeSimData[j][2]
    #print(row, 3*col, sim)
    if visited[row][col] == False :
      labThemeData.iloc[row,3*col] = L1
      labThemeData.iloc[row,3*col+1] = a1
      labThemeData.iloc[row,3*col+2] = b1
      recommandThemeOne.append(row)
      visited[row][col] = True
      count += 1
    if count >= 10 : 
      break
  recommandTheme = recommandTheme.append(pd.Series(recommandThemeOne, index=recommandTheme.columns), ignore_index=True)

labThemeData.loc[0]

labThemeData.to_csv('gdrive/My Drive/colorPOOL/kulerData/kulerLabData2.csv')
recommandTheme.to_csv('gdrive/My Drive/colorPOOL/kulerData/recommandTheme.csv')
```

```
def lab2rgb( l_s, a_s, b_s ) : 
  var_Y = ( l_s + 16. ) / 116.
  var_X = a_s / 500. + var_Y
  var_Z = var_Y - b_s / 200.

  if  pow(var_Y,3) > 0.008856 :
    var_Y = pow(var_Y,3)
  else :
    var_Y = ( var_Y - 16. / 116. ) / 7.787
  if  pow(var_X,3) > 0.008856 :
     var_X = pow(var_X,3)
  else :
    var_X = ( var_X - 16. / 116. ) / 7.787
  if  pow(var_Z,3) > 0.008856 :
    var_Z = pow(var_Z,3)
  else :
    var_Z = ( var_Z - 16. / 116. ) / 7.787

  X = 95.047 * var_X
  Y = 100.000 * var_Y   
  Z = 108.883 * var_Z    


  var_X = X / 100.
  var_Y = Y / 100.
  var_Z = Z / 100.

  var_R = var_X *  3.2406 + var_Y * -1.5372 + var_Z * -0.4986
  var_G = var_X * -0.9689 + var_Y *  1.8758 + var_Z *  0.0415
  var_B = var_X *  0.0557 + var_Y * -0.2040 + var_Z *  1.0570

  if  var_R > 0.0031308 :
    var_R = 1.055 * pow(var_R , ( 1 / 2.4 ))  - 0.055
  else : 
    var_R = 12.92 * var_R;
  if  var_G > 0.0031308 :
    var_G = 1.055 * pow(var_G , ( 1 / 2.4 ) )  - 0.055
  else :
    var_G = 12.92 * var_G
  if var_B > 0.0031308 :
    var_B = 1.055 * pow( var_B , ( 1 / 2.4 ) ) - 0.055
  else :
    var_B = 12.92 * var_B

  R = var_R * 255.
  G = var_G * 255.
  B = var_B * 255.

  return [R,G,B]
```
