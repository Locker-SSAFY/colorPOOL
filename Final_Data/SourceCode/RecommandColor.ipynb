```
# 피어슨 상관계수를 이용한 유사도 측정
# peason_matrix[x][y] : x와 y가 변하는 정도
# peason_matrix[x][y] 가 1에 가까우면 서로 강한 양적 관계(연관성이 높다)
# peason_matrix[x][y] 가 -1에 가까우면 서로 강한 음적 관계(연관성이 전혀 없다)
matrix = pd.read_csv('gdrive/My Drive/colorPOOL/kulerData/userColorData.csv')
matrix = matrix.iloc[:, 1:];
matrix
pearson_matrix = matrix.iloc[:,1:].corr(method='pearson')
pearson_matrix.to_csv('gdrive/My Drive/colorPOOL/kulerData/colorSimilarity.csv')
pearson_matrix
```

```
result = [[0,0,0,0,0]]
for i in range(len(pearson_matrix)) :
  max = -1.0
  max_index = 0
  arr = []
  temp = []
  for j in range(len(pearson_matrix)) :
    arr.append((pearson_matrix.iloc[i,j],j))
    if max < pearson_matrix.iloc[i,j] and pearson_matrix.iloc[i,j] <1.0 :
      max = pearson_matrix.iloc[i,j]
      max_index = j
  arr.sort(key=lambda x: x[0],reverse=True)
  for k in range(5) :
    temp.append(arr[k+1][1])
  result.append(temp)
  print(i ,max_index, max)
result
recommandColor = pd.DataFrame(data=result,columns=['color1','color2','color3','color4','color5'],dtype=int)
recommandColor.to_csv('gdrive/My Drive/colorPOOL/kulerData/recommandColor.csv')
recommandColor
```

```
### 이하는 코사인 유사도를 측정한 부분: 정확성이 떨어져서 피어슨 채택

# 코사인 유사도 공식
def cos_sim(A, B):
  normA = norm(A)
  normB = norm(B)
  if normA * normB == 0 : #부동 소수점 문제 해결
    return 0
  else :
    return dot(A, B)/(normA*normB)
```

```
# 기준 matrix의 유사도 구하기

arr = []
# 열 벡터를 기준으로 유사도 matrix 생성(열 : color 이므로)
for i in range(1,201) :
  temp = [] # 임시 저장 배열
  for j in range(1,201) :
    if i == j : 
      temp.append(1.0)
    else :  
      temp.append(cos_sim(matrix.iloc[:,i],matrix.iloc[:,j]))
  arr.append(temp)
sim_matrix = pd.DataFrame(data=arr,columns = [i for i in range(200)],dtype=float) 
sim_matrix.to_csv('gdrive/My Drive/colorPOOL/kulerData/colorSimilarity2.csv')
sim_matrix
```
