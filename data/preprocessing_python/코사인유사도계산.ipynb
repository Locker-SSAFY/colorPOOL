```
import pandas as pd
import numpy as np
from numpy import dot
from numpy.linalg import norm
```

```
def cos_sim(A, B):
  normA = norm(A)
  normB = norm(B)
  if normA * normB == 0 : #부동 소수점 문제 해결
    return 0
  else :
    return dot(A, B)/(normA*normB)
```

```
matrix = pd.read_csv("gdrive/My Drive/colorPOOL/kulerData/referenceMatrix.csv")
```

```
matrix = matrix.drop([matrix.columns[0]],axis=1);
```

```
matrix
```

```
cols_len = len(matrix.loc[0])
m_columns = matrix.columns

#similarity를 저장하는 행렬
sim_matrix = pd.DataFrame(columns=m_columns, dtype=float)

for r in range (0, len(matrix)): #len(matrix)
  max = 0
  max_index = -1
  for c in range(0, cols_len):
    rating = matrix.iloc[r][c]
    if rating:
      if max < rating:
        max = rating
        max_index = c
  
  kj = matrix.iloc[max_index]
  arr = [0] * cols_len
  for i in range(0, cols_len):
    if i!=max_index :
      ki = matrix.iloc[i]
      arr[i] = cos_sim(ki,kj)
    else :
      arr[i] = 1.0
  sim_matrix = sim_matrix.append(pd.Series(arr, index=m_columns), ignore_index=True)
  # temp_arr = []
  # for i in range(0, cols_len):
  #   if arr[i] !=0:
  #     temp_arr.append(arr[i])
  # print(temp_arr)
sim_matrix
sim_matrix.to_csv('gdrive/My Drive/colorPOOL/kulerData/simMatrix.csv')
```
